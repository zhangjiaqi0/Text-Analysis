a).	Write a python function that calculates the value of the function f above for any x. Do the same for the gradient f’.

```python
def f(x):
    """
    A function that calculate f(x) and returns 
    a continuous output
    """
    return(23+3*x+np.exp(-0.3*x)+0.5*x**2)
```

```python
def fg(x):
    """
    A function that calculate the gradient f(x) and returns 
    a continuous output
    """
    return(3-0.3*np.exp(-0.3*x)+x)
```

b). Plot the function and its gradient for values of x between -10 and 10. 

```python
x = np.linspace(-10, 10, 100)
y = f(x)
fig = plt.figure()
plop_f = fig.add_subplot(1,1,1)#显示方框的大小
plop_f.plot(x,y)
```

```python
yg = fg(x)
fig = plt.figure()#创建一个空白的画布
plop_fg = fig.add_subplot(1,1,1)#显示方框的大小
plop_fg.plot(x,yg)#Figure相当于一张白纸——可称做画布，Axes则是画布中的一块区域。Axes下面又包括了一下绘图较小的对象，如刻度线、图例和文本等
```

c).	Write a function that finds a minimum of the function f using the gradient descent algorithm described above. Your function should allow the user to choose any starting point x0 and learning rate η.

```python
def gradient_descent(x0, eta, tol = 1e-06, I =1000):
    """an algorithm that finds the value of x 
    which corresponds to the minimum of a function f(x)"""
    i = 0
    gap = 99999
    while gap > tol:#if the absolute value of x1-x0 is close to 0
        fg0 = fg(x0)
        x = x0 - eta * fg0
        fgx = fg(x)
        gap = abs(x -x0)
        if (gap > tol):
            x0 = x
            fg0 = fgx
        i += 1
        if (i > I):#if the current iteration is greater than the maximum I
            break
    return(x)
```

d).	Demonstrate that the function you wrote for part c) works by calling it and providing outputs for three randomly chosen starting points within the interval [-10,10], using a learning rate of η = 0.05. State the minimum of the function and indicate its location on a plot.

```python
import random


for i in range(3):
    eta = 0.05
    x = random.uniform(-10,10) #produce numbers randomly
    i += 1
    print("The minimum of the function = " + str(gradient_descent(x, eta)))
```
